<p>Like last time, this notebook is about extracting spatial data from ESRI shapefiles into Pandas DataFrames. Instead of using PyShp and Fiona, we'll use PySAL and OGR this time.</p>
<p><a href="http://pysal.readthedocs.org/en/latest/">PySAL</a> is an open source Python Spatial Analysis Library whose functionality goes far beyond, but includes, reading shapefiles. <a href="http://www.osgeo.org/gdal_ogr">OGR</a> is an open source library, written in C++, with python bindings that allow us to use its commands through the python language. OGR is probably the most widely used open source package for manipulating spatial data. It is used "under the hood" by software packages such as:
<ul><li>ESRI ArcMap</li>
    <li>Google Earth</li>
    <li>GeoDjango</li>
    <li>QGIS</li>
    <li>GRASS GIS... and more! <font size=2><a href="https://trac.osgeo.org/gdal/wiki/SoftwareUsingGdal">source</a></font></li>
</ul>
That being said, OGR isn't always the easiest to work with, even if it is quite powerful. Much of this post is spent working out how to extract the usual information (coordinates, geometry types, and attribute data) from shapefiles using OGR.</p>

<p>As always, if you simply want the final code, see directly below. There are three functions within one cell because using the "ogr" argument with the makeSpatialDF function calls the other two functions, get_type_and_coords() and make_tuples(), within the cell. The "pyshp", "fiona", and "pysal" arguments do not require any additional functions.</p>
<p>See below the cell for explanations/derivations.</p>

```python
def make_tuples(regex_coord_pairs):
    new_list = [] #Initialize a new list
    for pair in regex_coord_pairs: #Iterate over all pairs identified by the regular expression search
        new = pair.replace(" ",",") #Replace the separating spaces by commas
        new = "(" + new[:-1] + ")" + new[-1] #Make a string of the tuple by placing the object closing character outside the tuple
        new_list.append(new) #Add the representation to new_list
    return new_list

def get_type_and_coords(ogr_feat):
    ogr_geo = ogr_feat.geometry() #Get the geometry of the ogr feature
    well_known_string = ogr_geo.ExportToWkt() #Get the well-known text string for this feature's geometry
    new_string = deepcopy(well_known_string) #make a copy of the well-known text string
    new_string = new_string.replace("(", "[") #Change all the parenthesis to brackets to make a list
    new_string = new_string.replace(")", "]") #Change all the parenthesis to brackets to make a list
    space_loc = new_string.find(" ") #Find the first space in the string. This separates the geometry type from the coordinates
    geo_type = new_string[:space_loc] #Get the string representing the geometry type
    new_string = new_string[space_loc + 1:] #Isolate the coordinates
    
    #Find all coordinate pairs in the string
    coord_pair_list = re.findall("(?:-?[0-9]+\.?[0-9]* ?){1,3}[,)\]]", new_string)
    new_coord_pairs = make_tuples(coord_pair_list) #Convert all the coordinate pairs into string represntations of tuples
    #Create a dictionary relating the original coordinate pairs to the string representation of python tuples
    replace_dict = dict(zip(coord_pair_list, new_coord_pairs))
    for key in replace_dict: #Replace all the original coordinate pairs with their python tuple representations
        new_string = new_string.replace(key, replace_dict[key])
    geo_coords = ast.literal_eval(new_string) #convert the modified string to an actual python object
    return [geo_coords, geo_type]


def makeSpatialDF(path, method):
    """
    path = a path to a shapefile
    method = One of the strings, "fiona", "pyshp", "pysal", or 'ogr'. This specifies which module has been imported and should
    be used to create the dataframe.
    
    ====================
    Returns a dataframe of the spatial information and attributes associated with each feature in the shapefile at "path".
    """
    if method == "pyshp":
        _shpFile = shapefile.Reader(path) #Load the shapefile into pyshp
        attrs = [] #Create an empty array to hold the column headings from the attribute table
        for field in _shpFile.fields[1:]: #Iterate over the list of fields, ignore the "DeletionFlag" field
            attrs.append(field[0]) #Add the field name to the list column headings
        cols = [("geometry", "coordinates"), ("geometry", "type")] + attrs #Create list of dataframe columns
        data = [] #Create the overall array to hold all the data
        recs = _shpFile.shapeRecords() #Create a list of all the features/records in the shapefile
        for feat in recs: #Iterate over all features in the shapefile
            newRow = [] #Create a new array to hold each row's data
            newRow.append(feat.shape.__geo_interface__["coordinates"]) #Add the feature's coordinates to the row
            newRow.append(feat.shape.__geo_interface__["type"]) #Add the feature's geometry type to the row
            newRow.extend(feat.record) #Add the feature's attributes to the row
            data.append(newRow) #Add the row to the overall array of data
        return pd.DataFrame(data, columns = cols) #Create the dataframe from the data.
    elif method == "fiona":
        coll = list(fiona.open(path)) #Create the fiona collection of your shapefile
        attrs = coll[0]["properties"].keys() #Get the column headings from the attribute table
        cols = [("geometry", "coordinates"), ("geometry", "type")] + attrs #Create the dataframe columns
        data = [] #Create the overall array to hold all the data
        for feature in coll: #Iterate over all the features within the feature collection
            newRow = [] #Create a new array to hold each row's data
            newRow.append(feature["geometry"]["coordinates"]) #Add the feature's coordinates to the row
            newRow.append(feature["geometry"]["type"]) #Add the feature's geometry type to the row
            newRow.extend(feature["properties"].values()) #Add the feature's attributes to the row
            data.append(newRow) #Add the row to the overall array of data
        return pd.DataFrame(data, columns = cols) #Create the dataframe from the data.
    elif method == "pysal":
        dbf = path[:-3] + "dbf" #Create the path to the database file from the path to the shapefile
        db = pysal.open(dbf, "r") #Open the database file with pysal
        attributes = pd.DataFrame(db[:, :], columns = db.header) #Create a dataframe from the database's records
        features = list(pysal.open(path, "r")) #Create a list containing all the features in the shapefile
        coords = [] #Initialize an empty list for the coordinates of each feature
        types = [] #Initialize an empty list for the geometry type of each feature
        for x in features: #iterate over all the features
            coords.append(x.__geo_interface__["coordinates"]) #Add the feature's coordinates to coords
            types.append(x.__geo_interface__["type"]) #Add the feature's geometry type to types
        #Make a dataframe with two columns, one for the coordinates and one for the geometry type
        geoData = pd.DataFrame({("geometry", "coordinates"): coords, ("geometry", "type"): types})
        return geoData.join(attributes)
    elif method == "ogr":
        _shp = ogr.Open(path) #open the shapefile with ogr
        _lyr = _shp.GetLayer() #Access the shapefile's layer. Assumes that the shapefile has only one layer
        attr_cols = _lyr.GetFeature(0).keys() #Get a list of the shapefile attribute column names
        cols = [("geometry", "coordinates"), ("geometry", "type")] + attr_cols #List the dataframes columns
        recs = [] #Initialize an empty list for the shapefile records
        for i in range(_lyr.GetFeatureCount()): #Iterate over all of the shapefiles features
            feat = _lyr.GetFeature(i) #Get a particular feature.
            try:
                #Initialize this record's list with the feature's geometry type and geometry coordinates
                rec = get_type_and_coords(feat) 
            except Exception as inst: #If there is an error, print a useful error message
                print inst #Print the original error message
                raise Exception("Error on feature {}".format(i)) #Raise an error and print what feature had an error
            if rec[1] == "POINT" and isinstance(rec[0], list): #Check if the feature is a point with a list for coords
                rec[0] = rec[0][0] #If so, use the tuple--not a list of a tuple-- as the coordinates
            if rec[1] == "POINT" and len(rec[0]) == 3 and rec[0][2] == 0: #If the tuple has three coords and the third is zero
                rec[0] = (rec[0][0], rec[0][1]) #Reduce the point tuple to not show the unhelpful 0 coordinate
            for key in feat.keys(): #Iterate over the attributes of the feature
                rec.append(feat[key]) #Add the attributes to the list for the dataframe
            recs.append(rec) #Add the list for the given feature to the overall list of shapefile records
        return pd.DataFrame(recs, columns = cols) #Create the spatial dataframe
```

<p>Using the mechanisms above we have a way to access the attribute data for each of our features. Getting the spatial data for each feature (namely the coordinates) proved to be a harder task. What follows below is what seemed like the most straightforward way for me to get all of the coordinates for any type of feature (multi-polygon, multi-linestring, point, etc.) and to get the coordinates in the format used by <a href="http://geojson.org/geojson-spec.html">geojsons</a> and the geo-interface.</p>
<p>My approach was to 
    <ol><li>Export the geometry of the feature to well-known text which is very close to geojson format.</li>
        <li>Convert the well-known text into a perfect string representation of the desired python object using the <a href="https://docs.python.org/2/library/re.html#module-re">regular expressions library</a>.</li>
        <li>Convert the string representation from step 2 into a python object using the <a href="https://docs.python.org/2/library/ast.html">ast library</a>.</li>
    </ol>
</p>
<p>To be clear, there are probably other good ways to accomplish this task but this is what I thought of. Any suggestions on other ways of doing this are appreciated! (Especially since the ExportToWKT() method sometimes rounds its coordinates instead of retaining all siginificant digits)</p>

```python
geo = feat.geometry()
geo.ExportToWkt()
```

<p>The three things about the well-known text that have to be changed are that:
    <ol><li>The string contains the geometry type at the beginning of the string. The geometry type should be removed and appropriately stored.</li>
        <li>The well-known text uses parentheses to represent collections of objects. According to the geojson specification however, collections of objects need to be represented with arrays, and arrays use square brackets. All of the parentheses must therefore be changed to square brackets.</li>
        <li>The x-y-z coordinates of each point are only separated from each by a space and different points are only separated from each other by a comma. Following the convention adopted in the geo-interface, we'll contain each individual point in a tuple, and separate each tuple by a comma.</li>
    </ol>
</p>

<p>For all the details about the well-known text format, see page 51-62 of the Open Geospatial Consortium's <a href="http://www.opengeospatial.org/standards/is">OpenGIS Implementation Specification for Geographic information - Simple feature access - Part 1: Common architecture</a>
</p>

<p>The two functions below, get_type_and_coords() and make_tuples(), carry out the three steps outlined above. The way to conceptualize the use of the funtions is that a feature is passed to get_type_and_coords(). get_type_and_coords() will then call make_tuples() in the process of getting the job done.</p>

<p>The first thing that get_type_and_coords() will do is change all of the parenthesis to brackets. The second thing it does is separate the geometry type from the geometry coordinates. The last thing it does is place parentheses around all the points.</p>

<p>To place parentheses around all the points get_type_and_coords will use a regular expression to search for all the coordinates. Explaining the intricacies of how regex works is out of the scope of this post, but I learned everything used in this post from <a href="http://www.amazon.com/Mastering-Regular-Expressions-Jeffrey-Friedl/dp/0596528124">"Mastering Regular Expressions"</a> by Jeffrey Friedl. The way the particular expression in get_type and coords() works is as follows:
    <ul><li>The characters "(?:" signify that I'm creating a non-capturing group.</li>
        <li>The characters "-?[0-9]+\\.?[0-9]* ?" are a pattern that means look for a sequence that may or may not start with a dash, followed by one or more digits, possibly followed by a period, possibly followed by any number of digits possibly followed by a space.</li>
        <li>The characters "){1,3}" tell the regular expression library to close the group and search for 1 to 3 repetitions of the specified group.</li>
        <li>The  characters "[,)\\]]" tell the regrular expression to look for either a comma, closing parenthesis, or a closing bracket at the end of the one to three repetitions of the group.</li>
    </ul>
</p>

<p>After all of the coordinates are found, the spaces which separate the x, y, and z coordinates will be replaced with commas and the closing character (the comma, closing parenthesis, or closing bracket) will be placed after the closing parenthesis of the tuple. After this, the ast library will be used to convert the string into a python object</p>

```python
def make_tuples(regex_coord_pairs):
    new_list = [] #Initialize a new list
    for pair in regex_coord_pairs: #Iterate over all pairs identified by the regular expression search
        new = pair.replace(" ",",") #Replace the separating spaces by commas
        new = "(" + new[:-1] + ")" + new[-1] #Make a string of the tuple by placing the object closing character outside the tuple
        new_list.append(new) #Add the representation to new_list
    return new_list

def get_type_and_coords(ogr_feat):
    ogr_geo = ogr_feat.geometry() #Get the geometry of the ogr feature
    well_known_string = ogr_geo.ExportToWkt() #Get the well-known text string for this feature's geometry
    new_string = deepcopy(well_known_string) #make a copy of the well-known text string
    new_string = new_string.replace("(", "[") #Change all the parenthesis to brackets to make a list
    new_string = new_string.replace(")", "]") #Change all the parenthesis to brackets to make a list
    space_loc = new_string.find(" ") #Find the first space in the string. This separates the geometry type from the coordinates
    geo_type = new_string[:space_loc] #Get the string representing the geometry type
    new_string = new_string[space_loc + 1:] #Isolate the coordinates
    
    #Find all coordinate pairs in the string
    coord_pair_list = re.findall("(?:-?[0-9]+\.?[0-9]* ?){1,3}[,)\]]", new_string)
    new_coord_pairs = make_tuples(coord_pair_list) #Convert all the coordinate pairs into string represntations of tuples
    #Create a dictionary relating the original coordinate pairs to the string representation of python tuples
    replace_dict = dict(zip(coord_pair_list, new_coord_pairs))
    for key in replace_dict: #Replace all the original coordinate pairs with their python tuple representations
        new_string = new_string.replace(key, replace_dict[key])
    geo_coords = ast.literal_eval(new_string) #convert the modified string to an actual python object
    return [geo_coords, geo_type]
```

```python
def make_tuples(regex_coord_pairs):
    new_list = [] #Initialize a new list
    for pair in regex_coord_pairs: #Iterate over all pairs identified by the regular expression search
        new = pair.replace(" ",",") #Replace the separating spaces by commas
        new = "(" + new[:-1] + ")" + new[-1] #Make a string of the tuple by placing the object closing character outside the tuple
        new_list.append(new) #Add the representation to new_list
    return new_list

def get_type_and_coords(ogr_feat):
    ogr_geo = ogr_feat.geometry() #Get the geometry of the ogr feature
    well_known_string = ogr_geo.ExportToWkt() #Get the well-known text string for this feature's geometry
    new_string = deepcopy(well_known_string) #make a copy of the well-known text string
    new_string = new_string.replace("(", "[") #Change all the parenthesis to brackets to make a list
    new_string = new_string.replace(")", "]") #Change all the parenthesis to brackets to make a list
    space_loc = new_string.find(" ") #Find the first space in the string. This separates the geometry type from the coordinates
    geo_type = new_string[:space_loc] #Get the string representing the geometry type
    new_string = new_string[space_loc + 1:] #Isolate the coordinates
    
    #Find all coordinate pairs in the string
    coord_pair_list = re.findall("(?:-?[0-9]+\.?[0-9]* ?){1,3}[,)\]]", new_string)
    new_coord_pairs = make_tuples(coord_pair_list) #Convert all the coordinate pairs into string represntations of tuples
    #Create a dictionary relating the original coordinate pairs to the string representation of python tuples
    replace_dict = dict(zip(coord_pair_list, new_coord_pairs))
    for key in replace_dict: #Replace all the original coordinate pairs with their python tuple representations
        new_string = new_string.replace(key, replace_dict[key])
    geo_coords = ast.literal_eval(new_string) #convert the modified string to an actual python object
    return [geo_coords, geo_type]


def makeSpatialDF(path, method):
    """
    path = a path to a shapefile
    method = One of the strings, "fiona", "pyshp", "pysal", or 'ogr'. This specifies which module has been imported and should
    be used to create the dataframe.
    
    ====================
    Returns a dataframe of the spatial information and attributes associated with each feature in the shapefile at "path".
    """
    if method == "pyshp":
        _shpFile = shapefile.Reader(path) #Load the shapefile into pyshp
        attrs = [] #Create an empty array to hold the column headings from the attribute table
        for field in _shpFile.fields[1:]: #Iterate over the list of fields, ignore the "DeletionFlag" field
            attrs.append(field[0]) #Add the field name to the list column headings
        cols = [("geometry", "coordinates"), ("geometry", "type")] + attrs #Create list of dataframe columns
        data = [] #Create the overall array to hold all the data
        recs = _shpFile.shapeRecords() #Create a list of all the features/records in the shapefile
        for feat in recs: #Iterate over all features in the shapefile
            newRow = [] #Create a new array to hold each row's data
            newRow.append(feat.shape.__geo_interface__["coordinates"]) #Add the feature's coordinates to the row
            newRow.append(feat.shape.__geo_interface__["type"]) #Add the feature's geometry type to the row
            newRow.extend(feat.record) #Add the feature's attributes to the row
            data.append(newRow) #Add the row to the overall array of data
        return pd.DataFrame(data, columns = cols) #Create the dataframe from the data.
    elif method == "fiona":
        coll = list(fiona.open(path)) #Create the fiona collection of your shapefile
        attrs = coll[0]["properties"].keys() #Get the column headings from the attribute table
        cols = [("geometry", "coordinates"), ("geometry", "type")] + attrs #Create the dataframe columns
        data = [] #Create the overall array to hold all the data
        for feature in coll: #Iterate over all the features within the feature collection
            newRow = [] #Create a new array to hold each row's data
            newRow.append(feature["geometry"]["coordinates"]) #Add the feature's coordinates to the row
            newRow.append(feature["geometry"]["type"]) #Add the feature's geometry type to the row
            newRow.extend(feature["properties"].values()) #Add the feature's attributes to the row
            data.append(newRow) #Add the row to the overall array of data
        return pd.DataFrame(data, columns = cols) #Create the dataframe from the data.
    elif method == "pysal":
        dbf = path[:-3] + "dbf" #Create the path to the database file from the path to the shapefile
        db = pysal.open(dbf, "r") #Open the database file with pysal
        attributes = pd.DataFrame(db[:, :], columns = db.header) #Create a dataframe from the database's records
        features = list(pysal.open(path, "r")) #Create a list containing all the features in the shapefile
        coords = [] #Initialize an empty list for the coordinates of each feature
        types = [] #Initialize an empty list for the geometry type of each feature
        for x in features: #iterate over all the features
            coords.append(x.__geo_interface__["coordinates"]) #Add the feature's coordinates to coords
            types.append(x.__geo_interface__["type"]) #Add the feature's geometry type to types
        #Make a dataframe with two columns, one for the coordinates and one for the geometry type
        geoData = pd.DataFrame({("geometry", "coordinates"): coords, ("geometry", "type"): types})
        return geoData.join(attributes)
    elif method == "ogr":
        _shp = ogr.Open(path) #open the shapefile with ogr
        _lyr = _shp.GetLayer() #Access the shapefile's layer. Assumes that the shapefile has only one layer
        attr_cols = _lyr.GetFeature(0).keys() #Get a list of the shapefile attribute column names
        cols = [("geometry", "coordinates"), ("geometry", "type")] + attr_cols #List the dataframes columns
        recs = [] #Initialize an empty list for the shapefile records
        for i in range(_lyr.GetFeatureCount()): #Iterate over all of the shapefiles features
            feat = _lyr.GetFeature(i) #Get a particular feature.
            try:
                #Initialize this record's list with the feature's geometry type and geometry coordinates
                rec = get_type_and_coords(feat) 
            except Exception as inst: #If there is an error, print a useful error message
                print inst #Print the original error message
                raise Exception("Error on feature {}".format(i)) #Raise an error and print what feature had an error
            if rec[1] == "POINT" and isinstance(rec[0], list): #Check if the feature is a point with a list for coords
                rec[0] = rec[0][0] #If so, use the tuple--not a list of a tuple-- as the coordinates
            if rec[1] == "POINT" and len(rec[0]) == 3 and rec[0][2] == 0: #If the tuple has three coords and the third is zero
                rec[0] = (rec[0][0], rec[0][1]) #Reduce the point tuple to not show the unhelpful 0 coordinate
            for key in feat.keys(): #Iterate over the attributes of the feature
                rec.append(feat[key]) #Add the attributes to the list for the dataframe
            recs.append(rec) #Add the list for the given feature to the overall list of shapefile records
        return pd.DataFrame(recs, columns = cols) #Create the spatial dataframe
```

<font size=4 color="red">Note the results of using the "ogr" method in makeSpatialDF are not exactly equal to the results of using the other methods.</font><p>Most notably:
    <ol><li>The coordinates may sometimes have lower precision than with other methods based on how OGR exports the coordinates using the ExportToWKT() method. Examples include 53.1200006 being rounded to 53.12</li>
        <li>The coordinate type is all capitalized when using the ogr method</li>
        <li>Missing attributes are treated differently. In ogr, missing features often are nonetype objects since the  attribute would have been an empty string in the database file, and using ast.literal_eval() will convert empyt strings to "None" python objects.</li>
